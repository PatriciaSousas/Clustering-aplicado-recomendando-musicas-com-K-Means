## Clusterização por gênero - aplicando StandardScaler e o PCA

Vou importar o metodo pipeline do sklearn, esse metodo faz de forma automatica uma pipeline de machine learning 
onde se aplica sequencialmente uma lista de transformações ate um resultado final, em resumo uma gestão dos dados 

o StandardScaler fará uma padronização dos dados que eu vou usar  e o PCA fará uma redução de colunas aparti do valor que eu sugerir 
isso me ajuda a construir o padrao clusterizado 

```python
## construi minha pipeline denominando ela de pca,abro uma lista com os modulos que eu quero utiliza, chamo também o PCA e passo dois parametros
## o n_components que é a quantidade de explicacao que eu quero no algoritmo final, então de 11 colunas, aqui ele vai resumir pra 2, e o SEED que é importantes incluir 
## para os resultados do modelo de MI seja replicavél em qualquer maquina 

SEED = 1224

np.random.seed(1224)


pca= Pipeline ([('scaler', StandardScaler()), ('PCA', PCA (n_components =2, random_state = SEED))])
```
